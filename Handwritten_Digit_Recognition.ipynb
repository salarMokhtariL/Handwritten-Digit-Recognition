```python
# Imports 

import cv2                          # CV2 is about computer vision we need that to load img and process img
import numpy as np                  # Just work with Numpy arrays
import matplotlib.pyplot as plt     # Visualization of actual digits
import os
import tensorflow as tf             # tensorflow use for the ML part
```

```python
# Loading the Dataset

mnist = tf.keras.datasets.mnist    # Load directly from tensorflow we don't need to download CSV files and process them we just have it already in the module
```

```python
 # Train & Test

(x_trian, y_train), (x_test, y_test) = mnist.load_data()              # x-> pixel data & y-> classification
```

```python
# Normalizing 

x_trian = tf.keras.utils.normalize(x_trian, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
# pre-processing 

model = tf.keras.models.Sequential()
```

```python
# add some layer

model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))      # Flatten layer -> we flatten a certain input shape 
model.add(tf.keras.layers.Dense(128, activation='relu'))      # Dense layer -> It's just basic neural network layer that's connected or where each neuron is connected each other
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
# Compile the model

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the model

model.fit(x_trian, y_train, epochs=3)

model.save('handDigit.model')
```
